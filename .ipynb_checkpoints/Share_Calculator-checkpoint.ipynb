```python
import pandas as pd
import numpy as np
import random
from sklearn.metrics import (cohen_kappa_score, confusion_matrix, accuracy_score, 
                             precision_score, recall_score, f1_score, roc_auc_score)
```

```python
AGREEMENT_DF = pd.read_csv("Annotator_Agreement.csv")
```

```python
AGREEMENT_DF
```

```python
list(set(AGREEMENT_DF.label.to_list()))
```

```python
AGREEMENT_DF["case"] = [a.split("@")[0] for a in AGREEMENT_DF.case_label.to_list()]
```

```python
data = AGREEMENT_DF.rename(columns = {"17":"Annotator A", "22":"Annotator B", "23":"Annotator C", "ai":"Pseudo ground truth"})
```

```python
data
```

```python
#interstitial lung disease, pneumonia, pulmonary edema, pleural effusion, cardiomegaly, pneumothorax, and atelectasis
```

```python
def text_in(longer, shorters):
    for shorter in shorters:
        if shorter.lower() in longer.lower():
            return True, shorter
    return False, False
        
```

```python
target_labels = ["interstitial lung disease", "pneumonia", "pulmonary edema", "pleural effusion", "cardiomegaly", "pneumothorax", "atelectasis"]
Annotator_A = data["Annotator A"].to_list()
Annotator_B = data["Annotator B"].to_list()
Annotator_C = data["Annotator C"].to_list()
Pseudo_ground_truth = data["Pseudo ground truth"].to_list()
ai_prediction = data["ai_prediction"].to_list()
label = data["label"].to_list()
case = data["case"].to_list()
```

```python
new_df = {"Annotator A": [], "Annotator B": [], "Annotator C": [], "Malicious annotator A": [], "Malicious annotator B": [],"Malicious annotator C": [], "Malicious annotator D": [], "Malicious annotator E": [], "Pseudo ground truth": [], "ai_prediction":[], "case": [], "label": [], }
for idx, current_label in enumerate(label):
    logic , shorter = text_in(current_label, target_labels)
    if logic:
        new_df["Annotator A"].append(Annotator_A[idx])
        new_df["Annotator B"].append(Annotator_B[idx])
        new_df["Annotator C"].append(Annotator_C[idx])
        new_df["Malicious annotator A"].append(random.random() < 0.5)
        new_df["Malicious annotator B"].append(random.random() < 0.1)
        new_df["Malicious annotator C"].append(random.random() < 0.01)
        new_df["Malicious annotator D"].append(random.random() < 0.01)
        new_df["Malicious annotator E"].append(random.random() < 0.001)
        new_df["Pseudo ground truth"].append(Pseudo_ground_truth[idx])
        new_df["case"].append(case[idx])
        new_df["label"].append(shorter)
        new_df["ai_prediction"].append(ai_prediction[idx])
```

```python
new_df = pd.DataFrame(data = new_df)
```

```python
new_df.to_csv("agreement_list.csv", index= False)
```

```python
agr_df = pd.read_csv("agreement_list.csv")
```

```python
agr_df
```

```python
annotators = ["Annotator A", "Annotator B", "Annotator C", "Malicious annotator A", "Malicious annotator B", "Malicious annotator C", "Malicious annotator D", "Malicious annotator E"]
df_spec = {"label": [], "Annotator": [], "Accuracy": [], "Precision": [], "Recall": [], "F1": [], "AUC": [], "Cohen's kappa": [], "Prevalence": [], "True negative rate": [], "False positive rate": [], "True negative rate":[]}
for target_label in target_labels:
    print( f"Item: {target_label}")
    for annotator in annotators:
        human_annotation = agr_df[ agr_df["label"] == target_label ][annotator].to_list()
        AI_annotation = agr_df[ agr_df["label"] == target_label ]["Pseudo ground truth"].to_list()
        AI_prediction = agr_df[ agr_df["label"] == target_label ]["ai_prediction"].to_list()
        tn, fp, fn, tp = confusion_matrix(AI_annotation, human_annotation).ravel()
        
        accuracy = accuracy_score(AI_annotation, human_annotation)
        precision = precision_score(AI_annotation, human_annotation)
        recall = recall_score(AI_annotation, human_annotation)
        f1 = f1_score(AI_annotation, human_annotation)
        AUC = roc_auc_score(human_annotation, AI_prediction)
        cohen = cohen_kappa_score(AI_annotation, human_annotation)
        TNR = tn / (tn + fn)
        Prevalence = (tp+fn)/(tp+tn+fp+fn)
        FPR = fp / (fp + tn)
        TPR = tp / (tp + fn)
        print( f"{annotator} - pseudo GT" )
        print( f"cohen's kappa: {cohen}"  )
        print( f"True negative rate: {TNR}   Prevalence: {Prevalence}" )
        print( f"False positive rate: {FPR}     True positive rate: {TPR}   ")
        print( f"Accuracy : {accuracy}    Precision: {precision}    Recall: {recall}   F1: {f1}")
        print( f"AUC : {AUC}")
        df_spec["Annotator"].append(annotator)
        df_spec["label"].append(target_label)
        df_spec["Accuracy"].append(accuracy)
        df_spec["Precision"].append(precision)
        df_spec["Recall"].append(recall)
        df_spec["F1"].append(f1)
        df_spec["AUC"].append(AUC)
        df_spec["Cohen's kappa"].append(cohen)
        df_spec["True negative rate"].append(TNR)
        df_spec["Prevalence"].append(Prevalence)
        df_spec["False positive rate"].append(FPR)
        print()
```

```python
df_spec = pd.DataFrame(data = df_spec)
```

```python
df_spec
```

```python
df_spec["Precision / Prevalence ratio"] = df_spec["Precision"] / df_spec["Prevalence"]
```
